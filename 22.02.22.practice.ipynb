```python
def func_9(input_a, input_b):
    sum = input_a + input_b
    return sum

def func_10(input_c, input_d, input_e):
    sum_2 = func_9(input_c, input_d)
    result = sum_2 - input_e
    return result

print(func_10(1,2,3))
```

```python
def func_11(input_a, input_b):
    result_1 = input_a + input_b
    return result_1

def func_12(input_c, input_d):
    result_2 = input_c - input_d
    return result_2

def func_13(input_e, input_f, input_g):
    if input_g == "S" :
        result_3 = func_11(input_e, input_f)
        return result_3

    elif input_g =="D":
        result_4 = func_12(input_e, input_f)
        return result_4

    else :
        return "not matched"

print(func_13(1,2, "S"))
#none이 나오는 이유는 13번에 리턴값이 없어서
# 함수를 전역함수로 나눠 준 이유는 그러므로써 활용도가 더 높아진다. 
```

```python
def func_14(input_a, input_b):
    _list = []
    for i in range(input_a, input_b+1, 1): #마지막 값이 출력이 안되기 때문에 +1함
        if i % 2 == 0:
            _list.append(i)
    return _list

def func_15(input_a, input_b):
    _list = []
    for i in range(input_a, input_b+1, 1):
        if i % 2 == 1:
            _list.append(i)
    return _list

def func_16(input_a, input_b, input_c):

    if input_c == "짝수":
        return func_14(input_a, input_b)


    elif input_c == "홀수":
        return func_15(input_a, input_b)

print(func_16(1,100, "홀수"))
```

```python
import bisect

result = []
for score in [33, 99, 77, 70, 89, 90, 100]:
    pos = bisect.bisect_left([60, 70, 80, 90], score)  # 점수가 정렬되어 삽입될 수 있는 포지션을 반환
    grade = 'FDCBA'[pos]
    result.append(grade)

print(result)
```

```python
from graphlib import TopologicalSorter

ts = TopologicalSorter()

# 규칙1
ts.add('영어중급', '영어초급')  # 영어중급의 선수과목은 영어초급
ts.add('영어고급', '영어중급')  # 영어고급의 선수과목은 영어중급

# 규칙2
ts.add('영어문법', '영어중급')  # 영어문법의 선수과목은 영어중급
ts.add('영어고급', '영어문법')  # 영어고급의 선수과목은 영어문법

# 규칙3
ts.add('영어회화', '영어문법')  # 영어회화의 선수과목은 영어문법

print(list(ts.static_order()))  # 위상정렬한 결과를 출력
```
