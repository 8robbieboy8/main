```python
import pandas as pd

sales_df = pd. read_csv('./csv/Sales Records.csv')
sales_df.head()
```

```python
sales_df_2 = sales_df.drop(["Ship Date", "Order ID", 
                            "Sales Channel", "Total Revenue"], axis=1)
sales_df_2.head()
```

```python
sales_df_2.columns = ["지역", "국가", "상품타입",
                    "배송순위", "주문날짜", "판매개수",
                    "단가", "원가", "총비용", "총이윤"]

sales_df_2.head()
```

```python
sales_df_2.sort_values('국가', inplace=True)

sales_df_2.head()
```

```python
sales_df_2.reset_index(drop=True, inplace=True)
```

```python
sales_df_2.head()
```

```python
sales_df_2["총이윤"] = sales_df_2["총이윤"].astype(int) #소수점 제거하고, 정수로 만들기
```

```python
sales_df_2.head()
```

```python
sales_df_2["총비용"] = sales_df_2["총비용"].round(0).astype(int) 
#소수점 반올림 하는 방법
sales_df_2.head()
```

```python
sales_df_2["원가"] = sales_df_2["원가"].round(1) 
#소수점 하나 남기고 반올림
sales_df_2.head()
```

```python
sales_df_2["판매개수"].round(-1)
#음수를 사용하면 1자리를 버릴수 있다. 
```

```python
sales_df_2["주문날짜"] = pd.to_datetime(sales_df_2["주문날짜"],
                        format = "%m/%d/%Y") #y는 두자리 연도
sales_df_2.head()
```

```python
sales_df_2.groupby(["지역", "국가", "상품타입"]).mean()["총이윤"]
```

```python
sales_df_2.groupby(["지역", "국가", "상품타입"]).sum()["총이윤"]
```

```python
sales_df_2.groupby("국가").get_group("South Korea")
```

```python
sales_df_2["지역"].value_counts()

#df['country'].value_counts()
```

```python
sales_df_3 = sales_df_2[sales_df_2['국가'].isin(["South Korea", "China", "Togo"])]
sales_df_3.groupby('국가').mean()['총이윤']
```
